````
from sys import stdin


def maxSubArraySum(a, size):
    totalMax = -1000000000
    sum = 0
    start = 0
    end = 0
    temp = 0

    for i in range(0, size):
        sum += a[i]
        if totalMax < sum:
            totalMax = sum
            start = temp
            end = i

        if sum < 0:
            sum = 0
            temp = i + 1
    return start + 1, end + 1, totalMax


size = int(input())
array = [int(i) for i in input().split()]

start, end, total = maxSubArraySum(array, size)
print(start, end, total)
````
