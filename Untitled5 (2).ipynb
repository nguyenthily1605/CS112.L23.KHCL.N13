```python
import geopandas as gpd
import folium
import matplotlib.pyplot as plt
```

```python
!git clone https://github.com/CityScope/CSL_HCMC
```

```python
file = gpd.read_file("/content/CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_District_Level.shp")
```

```python
file.plot(column='Dist_Name', figsize=(16,8))
```

```python
m = folium.Map(location=[10.8, 106.7], zoom_start=10, tiles='CartoDB positron')
m
```

```python
for _, r in file.iterrows():
    sim_geo = gpd.GeoSeries(r['geometry']).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j,
                           style_function=lambda x: {'fillColor': 'orange'})
    folium.Popup(r['Dist_Name']).add_to(geo_j)
    geo_j.add_to(m)
  
m
```

```python
file = file.to_crs(epsg=2263)
file['centroid'] = file.centroid
file =file.to_crs(epsg=4326)
file['centroid'] = file['centroid'].to_crs(epsg=4326)
```

```python
for _, r in file.iterrows():
    lat = r['centroid'].y
    lon = r['centroid'].x
    folium.Marker(location=[lat, lon],
                  popup='length: {} <br> area: {}'.format(r['Shape_Leng'], r['Shape_Area'])).add_to(m)
```
